```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '###!', '####', '****', '*****', 'NULL']
df = pd.read_csv("ocorrencia.csv", sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        'codigo': pa.Column(pa.Int, required=False),
        'codigo_ocorrencia': pa.Column(pa.Int),
        'codigo_ocorrencia2': pa.Column(pa.Int),
        'ocorrencia_classificacao': pa.Column(pa.String),
        'ocorrencia_cidade': pa.Column(pa.String),
        'ocorrencia_uf': pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        'ocorrencia_aerodromo': pa.Column(pa.String, nullable=True),
        'ocorrencia_dia': pa.Column(pa.DateTime),
        'ocorrencia_hora': pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable=True),
        'total_recomendacoes': pa.Column(pa.Int)
        

    }
)
```

```python
schema.validate(df)
```

```python
filtro = df.ocorrencia_aerodromo.isnull()
df.loc[filtro]
```

```python
df.count() #faz a contagem de valores em cada coluna, n conta valores nulos
```

```python
#Ocorrencias com mais de 10 recomendacoes
filtro2 = df.total_recomendacoes > 10
df.loc[filtro2]
```

```python
#Quero agora ver a cidade e o total de recomendacções  
filtro2 = df.total_recomendacoes > 10
df.loc[filtro2, ['ocorrencia_cidade', 'total_recomendacoes']]
```

```python
#Ocorrencias com  classificação == incidente grave
filtro3 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro3]
```

```python
#Ocorrencias com  classificação == incidente grave  e o estado ==SP
filtro4 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro5= df.ocorrencia_uf == 'SP'
df.loc[filtro4 & filtro5]
```

```python
# Ocorrencias com  classificação == incidente grave ou incidente   e o estado ==SP
filtro6= df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro7= df.ocorrencia_uf == 'SP'
df.loc[filtro6 & filtro7]
```

```python
# Ocorrencias cuja cidade comecem com a letra C
filtro8= df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro8]
```

```python
# Ocorrencias cuja cidade terminam com a letra A
filtro9= df.ocorrencia_cidade.str[-1] == 'A'
df.loc[filtro9]
```

```python
# Ocorrencias cuja cidade terminam com MA
filtro10= df.ocorrencia_cidade.str[-2:] == 'MA'
df.loc[filtro10]
```

```python
# Ocorrencias cuja cidade tem a string MA
filtro11= df.ocorrencia_cidade.str.contains('MA')
df.loc[filtro11]
```

```python
# Ocorrencias cuja cidade tem a string MA ou AL
filtro12= df.ocorrencia_cidade.str.contains('MA|AL')
df.loc[filtro12]
```

```python
# Ocorrencias cujo ano foi 2015
filtro12= df.ocorrencia_dia.dt.year == 2015
df.loc[filtro12]
```

```python
# Ocorrencias cujo ano foi 2015 e mes 12

filtro13= df.ocorrencia_dia.dt.year == 2015
filtro14= df.ocorrencia_dia.dt.month == 12
filtro_dia= (df.ocorrencia_dia.dt.day > 2) & (df.ocorrencia_dia.dt.day < 9)

df.loc[filtro13 & filtro14 & filtro_dia] 

#filtro14= (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & (df.ocorrencia_dia.dt.day == 8)

```

```python
df['ocorrencia_dia_hora']= pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
df.head()
```

```python
# Ocorrencias cujo ano foi 2015 e mes 12

filtro13= df.ocorrencia_dia_hora.dt.year == 2015
filtro14= df.ocorrencia_dia_hora.dt.month == 12
filtro_dia= (df.ocorrencia_dia_hora.dt.day > 2) & (df.ocorrencia_dia.dt.day < 9)

df.loc[filtro13 & filtro14 & filtro_dia] 

#filtro14= (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12) & (df.ocorrencia_dia.dt.day == 8)

```

```python
filtro15= df.ocorrencia_dia_hora >= '2015-12-03 11:00:00'
filtro16= df.ocorrencia_dia_hora <= '2015-12-08 13:00:00'
df.loc[filtro15 & filtro16]
```

#### Agrupamento das ocorrencias no Sudeste em 2010

```python
#OCorrencias no sudeste no ano de 2010
filtro1= df.ocorrencia_dia.dt.year == 2010
filtro2= df.ocorrencia_uf.isin(['SP', 'MG', 'ES', 'RJ'])
df_sudeste=df.loc[filtro1 & filtro2]
df_sudeste
```

```python
df_sudeste.groupby(['ocorrencia_classificacao']).size()
```

```python
df_sudeste.groupby(['ocorrencia_uf','ocorrencia_classificacao']).size()
```

```python
# Quais cidades tiveram mais incidentes
df_sudeste.groupby(['ocorrencia_cidade']).size().sort_values(ascending=False)
```

```python
filtro_rj = df_sudeste.ocorrencia_cidade == 'RIO DE JANEIRO'
filtro2= df_sudeste.total_recomendacoes > 0
df_sudeste.loc[filtro_rj & filtro2]
```
